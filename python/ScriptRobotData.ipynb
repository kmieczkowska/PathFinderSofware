```python
file_path = os.path.join("data", "RobotDataTest1.csv")
csv_file = pd.read_csv(file_path)

# Konwertujemy 'nanoTime' (w milisekundach od epoch) na datę i czas
csv_file['timestamp'] = pd.to_datetime(csv_file['nanoTime'], unit='ms')

# Obliczamy różnicę czasu między kolejnymi wierszami w sekundach
csv_file['delta_seconds'] = csv_file['timestamp'].diff().dt.total_seconds()

# Wyświetlamy tylko potrzebne kolumny
csv_file[['nanoTime', 'timestamp', 'delta_seconds']].head(10)
```
