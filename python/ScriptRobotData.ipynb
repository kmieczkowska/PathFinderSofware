```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from collections import Counter
import numpy as np

# Wczytanie pliku CSV
csv_file = pd.read_csv(os.path.join("data", "RobotDataTest2.csv"))

# Zakładam, że kolumna z czasem to np. "time" i jest w sekundach lub ms
# Jeśli to ms, konwertujemy do sekund: csv_file["time"] // 1000
czasy_klatek = csv_file["time"] // 1  # jeśli są w sekundach
czasy_klatek = czasy_klatek.astype(int)  # upewniamy się, że są to inty

# Zliczanie FPS
fps_counter = Counter(czasy_klatek)
sekundy = sorted(fps_counter.keys())
fps = [fps_counter[s] for s in sekundy]

# Średnia liczba FPS
sredni_fps = np.mean(fps)
srednia_x = sekundy[len(sekundy) // 2]

# Wykres
plt.figure(figsize=(10, 6))
plt.bar(sekundy, fps, color='lightgreen', label='FPS w każdej sekundzie')
plt.scatter(srednia_x, sredni_fps, color='red', zorder=5, label=f'Średni FPS = {sredni_fps:.2f}')

plt.title('FPS na przestrzeni czasu')
plt.xlabel('Czas (s)')
plt.ylabel('Liczba klatek (FPS)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
