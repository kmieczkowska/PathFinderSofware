```python
def drawUsagePlot(data_with_filenames):
    """
    Plots Memory and CPU Usage over time for multiple datasets. Each file is plotted in a separate figure, 
    and all datasets are also plotted together in a combined plot.
    
    Args:
        data_with_filenames (list of tuples): Each tuple contains a filename and its corresponding DataFrame.
    """
    # Combined plot for all files
    plt.figure(figsize=(10, 8))
    for file_name, data in data_with_filenames:
        elapsed_time = data["Elapsed Time (s)"]
        memory_usage = data[" Memory usage Percentage"].round(2)

        print(memory_usage)
        
        cpu_usage = data["CPU Load"]
        
        # Plot Memory Usage
        plt.plot(elapsed_time, memory_usage, marker="o", label=f"{file_name} - Memory Usage (MB)")
        
        # # Plot CPU Usage
        # plt.plot(elapsed_time, cpu_usage, marker="s", label=f"{file_name} - CPU Usage (%)")
    
    plt.title("Memory and CPU Usage over Time (Combined Plot)")
    plt.xlabel("Elapsed Time (s)")
    plt.ylabel("Usage")
    plt.ylim(bottom=0)  # Adjust as needed
    plt.legend(title="File Names and Metrics")
    plt.grid()
    plt.show()
    
```

```python
# Define the folder containing the CSV files
folder_path = "data"

# Collect all CSV file paths in the folder
csv_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith(".csv")]

# Read each CSV file into a DataFrame and store them in a list
data_frames = [(os.path.basename(csv_file), pd.read_csv(csv_file)) for csv_file in csv_files]
```

```python
print("Rasperby sie zatrzymuje bo java kończy się pamięć (najprowdopodbniej bład w kodzie)")

drawFpsPlot(data_frames)
drawUsagePlot(data_frames)
```
